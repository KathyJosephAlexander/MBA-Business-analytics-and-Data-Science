Pseudocode of C4.5 algorithm

Let’s see the Pseudocode of the C4.5 algorithm in data mining.

    *   First, notice the base
    *   For each attribute X, find the normalized information gain ratio by splitting between X.
    *   Suppose that X is an attribute with the highest normalized information gain.
    *   Create a decision node that splits on attribute X.
    *   Repeat it on the sublists obtained by splitting the attribute X, and add these nodes as children of the node.


Advantages of C4.5 over other Decision Tree systems

    *   The algorithm is very helpful in Mitigating the overfitting because C4.5 inherently employs the Single Pass Pruning Process.
    *   C4.5 can work with Discrete data and can also work with Continuous Data
    *   C4.5  is very helpful in solving the issues of data incompleteness.

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, regressor.predict(x), color = 'blue')
plt.xlabel('Package')
plt.ylabel('Sales')
plt.show()
```
